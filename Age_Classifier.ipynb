```python
# prompt: funkion to upload foto and predict

from huggingface_hub import notebook_login
import requests
from PIL import Image
from io import BytesIO
from transformers import ViTFeatureExtractor, ViTForImageClassification
from google.colab import files

notebook_login()


def predict_age(uploaded_image):
    # Init model, transforms
    model = ViTForImageClassification.from_pretrained('nateraw/vit-age-classifier')
    transforms = ViTFeatureExtractor.from_pretrained('nateraw/vit-age-classifier')

    # Open the uploaded image
    im = Image.open(BytesIO(uploaded_image))

    # Transform our image and pass it through the model
    inputs = transforms(im, return_tensors='pt')
    output = model(**inputs)

    # Predicted Class probabilities
    proba = output.logits.softmax(1)

    # Predicted Classes
    preds = proba.argmax(1)

    # Assuming the model's output is a numerical age group
    # You'll likely need to map 'preds' to the actual age range
    # Example mapping (replace with the correct mapping for the model):
    age_mapping = {0: '0-10', 1: '10-20', 2: '20-30', 3: '30-40', 4: '40-50', 5: '50+'}
    predicted_age_group = age_mapping.get(preds.item(), "Unknown")

    return predicted_age_group


# Upload image
uploaded = files.upload()
for fn in uploaded.keys():
  image_bytes = uploaded[fn]
  predicted_age = predict_age(image_bytes)
  print(f"Predicted age group for {fn}: {predicted_age}")
```
